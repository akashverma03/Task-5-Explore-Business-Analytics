```python
sns.pairplot(df1)
plt.show()
```

```python
plt.figure(figsize=(30,30))
df1.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=df1)
plt.show()
```

```python
plt.figure(1)
plt.subplot(221)
df1['Category'].value_counts().plot.bar(figsize=(20,10), color='red', title='Category')
plt.subplot(222)
df1['Segment'].value_counts().plot.bar(color='blue', title='Segment')
plt.subplot(223)
df1['Ship Mode'].value_counts().plot.bar(color='green', title='Shop Mode')
plt.subplot(224)
df1['Region'].value_counts().plot.bar(color='pink', title='Region')

plt.show()
```

```python

plt.figure(figsize=(15,10))
df1['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
sns.scatterplot("Sales",'Profit',data=df1)
```

```python
pd.crosstab(df1["Quantity"],df1["Category"],df1["Profit"],aggfunc='sum').plot(kind="bar",stacked=True)
```

```python

sns.catplot(x="Sub-Category", y="Profit", data=df1, color='Green', height=4, aspect=4);
```

```python
df1.plot(kind='scatter', x='Quantity', y='Profit', figsize=(12, 6), color='orange')
plt.show()
```


From the above plot:
Profit is depending upon many other factors except quantity acting upon it.
